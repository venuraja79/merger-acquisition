```python
# Import the requirements
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
# Using the final data file 
file_name = "../Input Data/BreakUp 1 and 2 Output with Market Data for LR and ML vFinal.csv"
Y_column = "Acquisition Flag"
```

```python
def preprocess_col_header(df):
    default_cols = ['Company Name','Inc Year','Delete Flag','Acquisition Flag']
    col_1 = df.iloc[0].values
    col_1 = ["" if str(item) =='nan' else item for item in col_1]
    #col_1 = [item.split('-')[1] if '-' in col_1 else item for item in col_1]
    col_1 = [str(int(item.replace("Dec-", "")) + 2000) if 'Dec-' in str(item) else item for item in col_1]
    col_1 = [str(int(item.replace("-Dec", "")) + 2000) if '-Dec' in str(item) else item for item in col_1]
    col_1 = [f"{item}" for item in col_1]

    col_2 = df.iloc[1].values
    col_2 = ["" if str(item) =='NaN' else item for item in col_2]
    
    cols = []
    for a,b in zip(col_1, col_2):
        if str(b).strip() not in default_cols:
            cols.append(f"{a}-{b}")
        else:
            print(b)
            cols.append(b.strip())
    
    df.columns = cols
    df = df[2:]
    return df
```

```python
df = pd.read_csv(file_name, header=None)
# Last 6 columns are not required
#df = df.iloc[:,:-6]
df = preprocess_col_header(df).reset_index(drop=True)
df.head()
```

```python
print(f"Total companies : {df.shape[0]}")
print(f"Companies that have all data for all years : {df.dropna().shape[0]}")
```

```python
print(f"Total Acquired companies in the dataset : {df[df[Y_column]=='Yes'].shape[0]}")
```

```python
# TBAG50L80
flags_to_choose = ['TBAG80','TBAG50L80']
print(f"Total companies : {df[df['Delete Flag'].isin(flags_to_choose)].shape[0]}")
df_tba = df[df['Delete Flag'].isin(flags_to_choose)]
print(df_tba.shape)
df_tba.head()
```

```python
# We will drop the following columns
unique_cols_to_drop = ['Net Interest Margin (NIM) (%)','Market CaPrice to Earningsquity','Asset Turnover','EPS',
                       'Market Capitalization','No. of employees']

cols_to_drop = []
for year in range(2000, 2022):
    for col in unique_cols_to_drop:
        cols_to_drop.append(f"{str(year)}-{col}")
```

```python
df_tba.drop(columns=cols_to_drop, inplace=True)
```

```python
def fill_column(df):
    df_fill = df.interpolate(axis=1) #, limit_area='inside')
    return df_fill
```

```python
column_name = 'Return on equity'
year_all_cols = [f"{str(year)}-{column_name}" for year in range(2000, 2022)]
#year_all_cols
df_col = df_tba[year_all_cols]
df_col = df_col.astype(float)
```

```python
df_col = fill_column(df_col)
df_col.head()
```

```python
other_cols = ['Company Name','Inc Year','Delete Flag','Acquisition Flag']
```

```python
unique_cols = []
for idx, col in enumerate(df_tba.columns):
    if "2000-" in col:
        unique_cols.append(col.replace("2000-",""))
unique_cols
```

```python
df_filled = df_tba[other_cols]
for col in unique_cols:
    col_all_years = [f"{str(year)}-{col}" for year in range(2000, 2022)]
    df_col = df_tba[col_all_years]
    df_col = df_col.astype(float)
    df_col = fill_column(df_col)
    df_filled = pd.concat([df_filled, df_col], axis=1)
```

```python
def prepare_panel_data(df):
    df_tmp = pd.DataFrame()
    for index, dim in enumerate(unique_cols):
        cols = ["Company Name"]
        col_names = ["Company"]
        df_dim = pd.DataFrame()
        for i in range(2000, 2022):
            col_names.append(f"{i}")
            col_name = f"{i}-{dim}"
            cols.append(col_name)
        df_dim = df[cols]
        df_dim.columns = col_names
        df_dim["Dimension"] = dim
        melted = pd.melt(df_dim,
                  id_vars=['Company','Dimension'],
                  #value_vars=df.columns[[2,3]], 
                  var_name= 'year', 
                  value_name='value')
        df_tmp = pd.concat([df_tmp, melted])

    df_tmp = df_tmp.sort_values(['Company','Dimension'])
    #print(df_tmp.head(1))
    
    pivot = df_tmp.pivot_table(index=['Company', 'year'], columns='Dimension',
                     values='value', aggfunc='first').reset_index()
    pivot['year'] = pivot['year'].astype(int)
    pivot = pivot.set_index(['Company','year']).sort_index()
    return pd.DataFrame(pivot).reset_index()
```

```python
df_a = df_filled[df_filled[Y_column]=='Yes']
print(f"Acquired companies {df_a.shape[0]}")
df_a = prepare_panel_data(df_a)
#print(df_a.head(1))
df_a["Acquired"] = 1
```

```python
df_na = df_filled[df_filled[Y_column]=='No']
print(f"Non-Acquired companies {df_na.shape[0]}")
df_na = prepare_panel_data(df_na)
#print(df_na.head(1))
df_na["Acquired"] = 0
```

```python
df_panel = pd.concat([df_na, df_a])
df_panel.shape
```
